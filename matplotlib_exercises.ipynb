### #1. Use matplotlib to plot the following equation:

y = $x^2$ - x+2

- You'll need to write the code that generates the x and y points.

- Add an anotation for the point 0, 0, the origin.

````python
#create x datapoints
x =np.array(range(-50,50))

#create y datapoints
y = x**2 - (x+2)
````

````python
#create the plot
plt.plot(x,y)

#add title and labels
plt.title('y = $x^2$ - x+2')
plt.xlabel('x axis')
plt.ylabel('y axis')

#add annotation
plt.annotate('origin', xy= (0,0), xytext = (0, 500), arrowprops= {'arrowstyle':'fancy'})

#remove banner
plt.show()
````

<div class="alert alert-block alert-success">
<b>Remember:</b>

- <b>arrowprops = {}</b> gives an arrow pointing in that direction
- <b>xytext</b> shows where the text will be

### #2. Create and label 4 <u>separate</u> charts for the following equations (choose a range for x that makes sense)
- $y=\sqrt{x}$
- $y= x^3$
- $y=2^x$
- y=1/(x+1)

### $y=\sqrt{x}$

````python
#create data for x and y1 values
x= range(1,100)
y1 = [math.sqrt(n) for n in x]

#create the plot
plt.plot(x,y1)

#add title and axis labels
plt.title('y=√x')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
````

<hr style="border:1px solid grey">

###  $y=x^3$

````python
#create data for x and y2 values
x= range(1,100)
y2 = [n **3 for n in x]

#create the plot
plt.plot(x,y2)

#add title and axis labels
plt.title('y = $x^3$')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
````

<hr style="border:1px solid grey">

### $y=2^x$

````python
#create data for x and y3 values
x= range(1,100)
y3 = [(2 **n) for n in x]

#create the plot
plt.plot(x,y3)

#add title and axis labels
plt.title('y = $2^x$')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show() 
````

<hr style="border:1px solid grey">

### y=1/(x+1)

````python
#create data for x and y4 values
x= range(1,100)
y4= [(1/(n+1)) for n in x]

#create the plot
plt.plot(x,y4)

#add title and axis labels
plt.title('y = $1/(x+1)$')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
````

### #3. Combine the figures you created in the last step into one large figure with 4 subplots.

````python
x= np.array(range(100))
y1 = [math.sqrt(x) for x in x]
y2 = [x **3 for x in x]
y3 = [2**x for x in x]
y4= [1/(x+1) for x in x]
````

````python
plt.figure(figsize = (12,10))

plt.suptitle('Separate Charts')

plt.subplot(221)
plt.plot(x,y1, color = 'orange')
plt.title('y1 = $√x$')

plt.subplot(222)
plt.plot(x,y2, color = 'g') 
plt.title ('y2 = $x^3$')


plt.subplot(223)
plt.plot(x,y3, color = 'm') 
plt.title ('y3 = $2^x$')


plt.subplot(224)
plt.plot(x,y4, color= 'black') 
plt.title ('y4 = $1/(x+1)$')

plt.tight_layout()

plt.show()
````

### Another way to accomplish this!

<b>Using OOP with Matplotlib</b> (object oriented plotting)

````python
#plt.subplots(rows, columns)
fig, axs = plt.subplots(2,2)

#axs index does start at 0
#row 1, column 1
axs[0,0].plot(x, y1)

#row 1, column 2
axs[0,1].plot(x, y2)

#row 2, column 1
axs[1,0].plot(x,y3)

#row 2, column 2
axs[1,1].plot(x,y4)
````

<div class="alert alert-block alert-success">
<b>NOTE:</b>
    
<b>Pyplot vs OOP Matplotlib</b>

```plt.xlabel() → ax.set_xlabel()```  
```plt.ylabel() → ax.set_ylabel()```
    
```plt.xlim() → ax.set_xlim()``` 
    
```plt.ylim() → ax.set_ylim()```
    
```plt.title() → ax.set_title()``` 
</div>

### #4. Combine the figures you created in the last step into one figure 
- where each of the 4 equations has a different color for the points. 
- Be sure to include a legend and an appropriate title for the figure.

````python
x= list(range(10))
y1 = [math.sqrt(x) for x in x]
y2 = [x **3 for x in x]
y3 = [(2 **n) for n in x]
y4= [(1/(n+1)) for n in x]
````

````python
#determine figure size
plt.figure(figsize =(10,8))

#add axis labels and title
plt.title('Different Equations Zoomed In')
plt.xlabel('x axis')
plt.ylabel('y axis')

#plot all 4 lines
plt.plot(x, y1, color = 'orange', label = 'y1 = $√x$')
plt.plot(x, y2, color = 'green', label = 'y2 = $x^3$')
plt.plot(x, y3, color = 'm', label = 'y3 = $2^x$')
plt.plot(x, y4, color = 'black', label = 'y4 = $1/(x+1)$')

#add gridlines
plt.grid(True, ls= '--')

#add legend
plt.legend(loc = 'upper right')

plt.ylim(0,8)
plt.xlim(0,8)

plt.show()
````

# Big O Notation

### #5. Make a new Jupyter notebook named big_o_notation.ipynb

- Title your chart "Big O Notation"
- Label your x axis "Elements"
- Label your y axis "Operations"
- Label your curves or make a legend for the curves
- Use LaTex notation where possible 
<br>

Curves to graph:
- y =0n+1 and label the curve "O(1)"
- y = log(n) and label the curve "O(log n)"
- y = n and label the curve "O(n)"
- y = n*log(n) and label it "O(n log n)"
- y = $n^2$ and label it "O(n^2)"
- y = $2^n$ and label it "O(2^n)"
- y= n! and label it "O(n!)"
- y= $n^n$ and label it "O(n^n)"

````python
x= (range(1,20))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4= [(n * math.log(n)) for n in x]
y5= [n**2 for n in x]
y6= [2**n for n in x]
y7= [math.factorial(n) for n in x]
y8= [n**n for n in x]

plt.figure(figsize = (16,8))

plt.plot(x,y1, color = 'cyan', label= 'y1= $O(1)$', alpha = .2)
plt.plot(x,y2, color = 'red', label= 'y2 = $O(log n)$', alpha = .3)
plt.plot(x,y3, color = 'b', label= 'y3= $O(n)$', alpha = .4)
plt.plot(x,y4, color= 'g', label= 'y4 = $O(n log n)$', alpha = .5)
plt.plot(x,y5, color= 'coral', label= 'y5 = $O(n^2)$', alpha = .6)
plt.plot(x,y6, color= 'olive', label= 'y6 = $O(2^n)$', alpha = .7)
plt.plot(x,y7, color= 'lime', label= 'y7 = $O(n!)$', alpha = .8)
plt.plot(x,y8, color= 'violet', label= 'y8 = $O(n^n)$', alpha = .9)

plt.title('Big O Notation', fontsize = 20)
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend(loc = 'upper right', fontsize = 14)
#plt.xlim(1, 11)
#plt.ylim(0, 50)

plt.show()
````
